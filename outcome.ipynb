```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def plot_outcome(file_1: str, file_2: str):
    utility_list = []
    with open(file_1) as f:
        reader = csv.reader(f)
        for row in reader:
            # print(row[0])
            utility_list.append(float(row[0]))


    utility_list_other = []
    with open(file_2) as f:
        reader = csv.reader(f)
        for row in reader:
            # print(row[0])
            utility_list_other.append(float(row[0]))



    plt.hist(utility_list, bins=100, alpha=0.8, label='Utilities',color="red")


    print("np.mean(utility_list): ",np.mean(utility_list))

    plt.hist(utility_list_other, bins=100, alpha=0.8, label='Other Value',color="blue")

    plt.legend()

    plt.show()

    print("np.mean(all_utility_list_other): ",np.mean(utility_list_other))
```

```python
plot_outcome('outcome/outcome_p.csv','outcome/outcome_pq.csv')
```

```python
plot_outcome('outcome/outcome_p_10_100.csv','outcome/outcome_pq_10_100.csv')
```
