```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def plot_outcome(file_1: str, file_2: str):
    utility_list = []
    with open(file_1) as f:
        reader = csv.reader(f)
        for row in reader:
            # print(row[0])
            utility_list.append(float(row[0]))


    utility_list_other = []
    with open(file_2) as f:
        reader = csv.reader(f)
        for row in reader:
            # print(row[0])
            utility_list_other.append(float(row[0]))


    print("Utility_list")
    print("Mean: ",np.mean(utility_list))
    print("Variance: ",np.var(utility_list))
    print("----")
    print("Utility_list_other")
    print("Mean: ",np.mean(utility_list_other))
    print("Variance: ",np.var(utility_list_other))


    plt.hist(utility_list, bins=100, alpha=0.8, label='Utilities',color="red")

    plt.hist(utility_list_other, bins=100, alpha=0.8, label='Other Value',color="blue")

    plt.legend()

    plt.show()

    
```

```python
plot_outcome('outcome/outcome_p_4_3_15.csv','outcome/outcome_pq_4_3_15.csv')
```
